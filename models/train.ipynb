```python
train = pd.read_csv("/Users/parksung-cheol/show-gy-AI/dacon-project/data/raw/train.csv")
test = pd.read_csv("/Users/parksung-cheol/show-gy-AI/dacon-project/data/raw/test.csv")
```

```python
"""generation, class4, contest_award, previous_class_3, completed_semester, incumbents_lecture_scale_reason, interested_company, contest_participation, idea_contest"""
```

```python
columns_to_drop = ['generation', 'class4', 'contest_award', 'previous_class_3', 'completed_semester', 
                   'incumbents_lecture_scale_reason', 'interested_company', 'contest_participation', 'idea_contest']
train = train.drop(columns=columns_to_drop)
print(f"삭제 후 train shape: {train.shape}")
```

```python
"""결측치 null로 채우기"""
```

```python
# 결측치 확인
print("결측치 개수:")
print(train.isnull().sum())
print(f"\n전체 결측치: {train.isnull().sum().sum()}")
print(f"\n결측치 비율:\n{train.isnull().sum() / len(train) * 100}")
```

```python
# 결측치를 적절한 값으로 채우기
# 수치형 컬럼: 중앙값으로 채우기
# 범주형 컬럼: 최빈값 또는 'Unknown'으로 채우기

for column in train.columns:
    if train[column].isnull().sum() > 0:
        if train[column].dtype in ['int64', 'float64']:
            # 수치형은 중앙값으로
            train[column].fillna(train[column].median(), inplace=True)
        else:
            # 범주형은 최빈값으로 (없으면 'Unknown')
            mode_value = train[column].mode()
            if len(mode_value) > 0:
                train[column].fillna(mode_value[0], inplace=True)
            else:
                train[column].fillna('Unknown', inplace=True)

print("결측치 처리 완료!")
print(f"남은 결측치: {train.isnull().sum().sum()}")
```

```python
# 모든 컬럼을 category형으로 변환 (completed 제외)
# float 값을 string으로 먼저 변환 후 category로 변환
for column in train.columns:
    if column != 'completed':
        # float나 int 값을 모두 string으로 변환 후 category로
        train[column] = train[column].astype(str).astype('category')
    else:
        train[column] = train[column].astype('int')

print("컬럼 타입 변환 완료!")
print(f"\n데이터 타입 확인:")
print(train.dtypes)
```

```python
# completed의 분포 확인
print("오버샘플링 전:")
print(train['completed'].value_counts())
print(f"비율: {train['completed'].value_counts(normalize=True)}")

# completed가 1인 행만 추출
completed_1 = train[train['completed'] == 1]

# completed가 1인 행을 복제해서 train에 추가
train_oversampled = pd.concat([train, completed_1], ignore_index=True)

# 데이터 셔플
train_oversampled = train_oversampled.sample(frac=1, random_state=42).reset_index(drop=True)

print(f"\n오버샘플링 후:")
print(train_oversampled['completed'].value_counts())
print(f"비율: {train_oversampled['completed'].value_counts(normalize=True)}")
print(f"\n데이터 shape: {train_oversampled.shape}")

# train을 오버샘플링된 데이터로 교체
train = train_oversampled
```

```python
from catboost import CatBoostRegressor
import numpy as np
```

```python
# ID 열 제외하고 X, y 분리
X = train.drop(['ID', 'completed'], axis=1)
y = train['completed']

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
print(f"\n특성 목록:\n{X.columns.tolist()}")
```

```python
# 각 칼럼별로 completed의 평균을 구해서 초기값 설정
# 예: True일 때 completed 평균이 0.75라면, True인 모든 행에 0.75를 할당

baseline_values = np.zeros(len(train))

for col in X.columns:
    # 각 카테고리별 completed 평균 계산
    category_means = train.groupby(col)['completed'].mean()
    
    print(f"\n{col} 카테고리별 평균:")
    print(category_means)
    
    # 각 행에 해당하는 카테고리의 평균값을 초기값으로 설정
    for idx in range(len(train)):
        category_value = train[col].iloc[idx]
        baseline_values[idx] += category_means[category_value]

# 모든 컬럼의 평균을 더했으므로, 컬럼 개수로 나눠서 평균 초기값 생성
baseline_values = baseline_values / len(X.columns)

print(f"\n초기값 통계:")
print(f"평균: {baseline_values.mean():.4f}")
print(f"최소: {baseline_values.min():.4f}")
print(f"최대: {baseline_values.max():.4f}")
```

```python
# CatBoost 모델 설정 및 학습
# 과적합을 강하게 유도하기 위한 설정

# 카테고리형 특성 인덱스 지정
cat_features = list(range(len(X.columns)))

model = CatBoostRegressor(
    iterations=5000,           # 반복 횟수 대폭 증가
    depth=10,                  # 트리 깊이 (과적합 유도)
    learning_rate=0.05,        # 학습률 약간 증가
    loss_function='RMSE',      # 손실 함수
    cat_features=cat_features, # 카테고리형 특성
    l2_leaf_reg=1,             # L2 정규화 약하게 설정
    random_seed=42,
    verbose=200                # 200번마다 출력
)

# 초기값(baseline)을 사용하여 학습
print("모델 학습 시작...")
model.fit(
    X, y,
    baseline=baseline_values,  # 각 칼럼별 평균을 초기값으로 사용
    verbose=200
)

print("\n모델 학습 완료!")
```

```python
# 학습 결과 확인
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score

y_pred = model.predict(X)

# RMSE, R2 Score
rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)

# Accuracy: 0.5 기준으로 0/1 분류
y_pred_binary = (y_pred >= 0.5).astype(int)
accuracy = accuracy_score(y, y_pred_binary)

print(f"Train RMSE: {rmse:.4f}")
print(f"Train R2 Score: {r2:.4f}")
print(f"Train Accuracy: {accuracy:.4f} ({accuracy*100:.2f}%)")

# 특성 중요도 확인
feature_importance = model.get_feature_importance()
feature_names = X.columns

importance_df = pd.DataFrame({
    'feature': feature_names,
    'importance': feature_importance
}).sort_values('importance', ascending=False)

print(f"\n특성 중요도 Top 10:")
print(importance_df.head(10))
```
