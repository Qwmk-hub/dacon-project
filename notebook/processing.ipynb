```python
import pandas as pd

# 1. 생성할 목표 칼럼 리스트 (요청하신 순서대로 정의)
target_columns = [
    '복수 전공 ( 다중전공, 이중전공 포함 )', 
    '단일 전공', 
    '단일 전공공학 (컴퓨터 공학 제외)', 
    '경제통상학', 
    '자연과학', 
    '예체능', 
    '사회과학', 
    'IT(컴퓨터 공학 포함)', 
    '경영학', 
    '의약학', 
    '교육학', 
    '인문학', 
    '법학', 
    '기타'
]

# 2. major_field 열은 '자연과학, 사회과학' 처럼 콤마로 구분되어 있으므로 미리 분리
#    (분리 후에는 해당 전공 이름의 칼럼이 생기고 값은 1이 됨)
field_dummies = major['major_field'].str.get_dummies(sep=', ')

# 3. 결과를 담을 DataFrame 생성 (ID 열 미리 포함)
major_information = pd.DataFrame()
major_information['ID'] = major['ID']

# 4. 루프를 돌면서 각 목표 칼럼에 해당하는 값이 데이터에 있는지 확인
for col in target_columns:
    # (1) major type 열에 해당 값이 있는지 확인
    cond1 = (major['major type'] == col)
    
    # (2) major1_1 열에 해당 값이 있는지 확인
    cond2 = (major['major1_1'] == col)
    
    # (3) major1_2 열에 해당 값이 있는지 확인
    cond3 = (major['major1_2'] == col)
    
    # (4) major_field (콤마로 분리한 것)에 해당 값이 있는지 확인
    if col in field_dummies.columns:
        cond4 = (field_dummies[col] == 1)
    else:
        cond4 = False # field_dummies에 해당 칼럼이 아예 없으면 False
        
    # 위 4가지 조건 중 하나라도 만족하면 1, 아니면 0 (OR 연산)
    # astype(int)를 사용하여 True/False를 1/0으로 변환
    major_information[col] = (cond1 | cond2 | cond3 | cond4).astype(int)
```

```python
import pandas as pd

# 1. 목표로 하는 반 번호 리스트 (요청하신 순서대로)
target_nums = [1, 2, 4, 5, 6, 7, 8, 11, 12, 13]

# 2. 결과를 담을 DataFrame 생성 (ID 포함)
class_information = pd.DataFrame()
class_information['ID'] = classs['ID']

# 3. 데이터를 확인할 컬럼 목록 (class1 ~ class4)
cols_to_check = ['class1', 'class2', 'class3', 'class4']

# 4. 각 번호에 대해 루프를 돌며 새로운 열 생성
for num in target_nums:
    # 새로운 열 이름 생성 (예: '1반', '2반'...)
    col_name = f"{num}반"
    
    # (1) .eq(num): 데이터프레임 전체에서 num과 같은 값을 찾습니다 (4.0 == 4 로 자동 인식)
    # (2) .any(axis=1): 가로(행) 방향으로 하나라도 True가 있으면 True를 반환합니다.
    # (3) .astype(int): True/False를 1/0으로 변환합니다.
    class_information[col_name] = classs[cols_to_check].eq(num).any(axis=1).astype(int)
```

```python
import pandas as pd

# 1. 생성할 목표 강의 리스트 (요청하신 문자열 그대로 사용)
target_courses = [
    "0001 : 파이썬 문법 기초반",
    "0002 : SQL 문법 기초반",
    "0003 : R문법 기초반",
    "0004 : 데이터 분석 입문반 (머신러닝)",
    "0005 : 데이터 분석 입문반 (통계)",
    "0006 : 데이터 분석 전처리반(판다스)",
    "0007 : 데이터 분석 전처리 적용반",
    "0008 : 데이터 분석 모델링반 (ML1)",
    "0009 : 데이터 분석 개념반"
]

# 2. 결과를 담을 DataFrame 생성 (ID 포함)
previous_information = pd.DataFrame()
previous_information['ID'] = previous['ID']

# 3. 탐색할 데이터 컬럼 지정 (ID를 제외한 나머지 모든 컬럼)
cols_to_check = [col for col in previous.columns if col != 'ID']

# 4. 문자열 검색을 위해 NaN 값을 빈 문자열('')로 임시 변환
temp_df = previous[cols_to_check].fillna('')

# 5. 각 강의에 대해 루프를 돌며 1/0 컬럼 생성
for course in target_courses:
    # (1) 모든 컬럼을 돌면서 해당 강의 문자열이 포함되어 있는지 확인 (regex=False로 괄호 등 특수문자 그대로 인식)
    # (2) 하나라도 포함되어 있으면 True가 되도록 OR(|) 연산 누적
    has_course = pd.Series([False] * len(previous), index=previous.index)
    for col in cols_to_check:
        has_course |= temp_df[col].str.contains(course, regex=False)
    
    # (3) True/False를 1/0으로 변환하여 할당
    previous_information[course] = has_course.astype(int)
```

```python
import pandas as pd

# 1. 결과를 담을 DataFrame 생성 (ID 포함)
why_encoded = pd.DataFrame()
why_encoded['ID'] = why['ID']

# 문자열 검색을 위해 NaN을 빈 문자열로 처리
why_filled = why.fillna('')

# ---------------------------------------------------------
# Group 1. 유입 경로 (inflow_route)
# ---------------------------------------------------------
inflow_cols = [
    '에브리타임', 
    '지인 추천', 
    '인스타그램', 
    '기존 학회원 또는 운영진',
    '대외활동 사이트(링커리어, 캠퍼스픽, 캠퍼즈, 위비티 등)', 
    '교내 플랫폼 (에브리타임 X)', 
    '인터넷 검색'
]

for col in inflow_cols:
    # 텍스트가 포함되어 있으면 1, 아니면 0 (괄호 등 특수문자 그대로 인식하기 위해 regex=False)
    why_encoded[col] = why_filled['inflow_route'].str.contains(col, regex=False).astype(int)

# '유입기타'는 '기타'라는 값이 있을 때 1로 설정
why_encoded['유입기타'] = why_filled['inflow_route'].eq('기타').astype(int)


# ---------------------------------------------------------
# Group 2. 지원 동기 (whyBDA)
# ---------------------------------------------------------
why_bda_cols = [
    '큰 규모인 만큼, 커리큘럼이나 운영 등 관리가 잘 될것 같아서',
    'BDA 학회원만의 혜택을 누리고 싶어서(현직자 강연, 잡 페스티벌, 기업연계 공모전 등)',
    '혼자 공부하기 어려워서',
    '이전 기수에 매우 만족해서',
    '시간적으로 부담이 없어서',
    '학회 가입 시 코딩 테스트, 면접 등을 보지 않아서',
    '현직자의 강의를 듣고 싶어서'
]

for col in why_bda_cols:
    why_encoded[col] = why_filled['whyBDA'].str.contains(col, regex=False).astype(int)


# ---------------------------------------------------------
# Group 3. 얻어가고 싶은 것 (what_to_gain)
# ---------------------------------------------------------
# 1) 일반 항목
gain_cols = [
    '프로젝트 경험', 
    '공모전 경험', 
    '인적 네트워크', 
    '기초적인 python, sql 지식 숙지'
]
for col in gain_cols:
    why_encoded[col] = why_filled['what_to_gain'].str.contains(col, regex=False).astype(int)

# 2) 동의어 처리: 데이터 분석 역량 (= “분석 역량”)
# "데이터 분석 역량" 혹은 "분석 역량" 둘 중 하나라도 포함되면 1
cond_da = (why_filled['what_to_gain'].str.contains('데이터 분석 역량', regex=False) | 
           why_filled['what_to_gain'].str.contains('분석 역량', regex=False))
why_encoded['데이터 분석 역량'] = cond_da.astype(int)


# ---------------------------------------------------------
# Group 4. 희망 도메인 (expected_domain)
# ---------------------------------------------------------
domain_cols = [
    'A. 농업, 임업 및 어업', 'C. 제조업', 'D. 전기, 가스, 증기 및 공기 조절 공급업',
    'E. 수도, 하수 및 폐기물 처리, 원료 재생업', 'F. 건설업', 'G. 도매 및 소매업',
    'H. 운수 및 창고업', 'I. 숙박 및 음식점업', 'J. 정보통신업', 'K. 금융 및 보험업',
    'L. 부동산업', 'M. 전문, 과학 및 기술 서비스업', 'N. 사업시설 관리, 사업 지원 및 임대 서비스업',
    'O. 공공 행정, 국방 및 사회보장 행정', 'P. 교육 서비스업', 'Q. 보건업 및 사회복지 서비스업',
    'R. 예술, 스포츠 및 여가관련 서비스업', 'S. 협회 및 단체, 수리 및 기타 개인 서비스업',
    'T. 가구 내 고용활동 및 달리 분류되지 않은 자가 소비 생산활동', 'U. 국제 및 외국기관'
]

for col in domain_cols:
    why_encoded[col] = why_filled['expected_domain'].str.contains(col, regex=False).astype(int)


# ---------------------------------------------------------
# Group 5. 원데이 클래스 주제 (onedayclass_topic)
# ---------------------------------------------------------
oneday_cols = [
    'Python 응용', 'SQL 응용', '데이터 시각화 (Matplotlib, Seaborn 등)',
    '머신러닝 / 딥러닝 응용', '웹 크롤링', '금융데이터', '금융공학',
    '데이터 엔지니어링(파이프라인설계)', '포트폴리오 A to Z', '공모전 수상 tip',
    '구글 애널리틱스', 'ChatGPT 활용', 'CNN', 'RNN'
]
for col in oneday_cols:
    why_encoded[col] = why_filled['onedayclass_topic'].str.contains(col, regex=False).astype(int)

# 동의어 처리 1: 태블로 (표기 “테블로” 포함)
cond_tableau = (why_filled['onedayclass_topic'].str.contains('태블로', regex=False) | 
                why_filled['onedayclass_topic'].str.contains('테블로', regex=False))
why_encoded['태블로 (표기 “테블로” 포함)'] = cond_tableau.astype(int)

# 동의어 처리 2: Power BI (표기 “Power Bi”, “파워비아이”, “powebi” 포함)
cond_pbi = (why_filled['onedayclass_topic'].str.contains('Power BI', regex=False) | 
            why_filled['onedayclass_topic'].str.contains('Power Bi', regex=False) |
            why_filled['onedayclass_topic'].str.contains('파워비아이', regex=False) |
            why_filled['onedayclass_topic'].str.contains('powebi', regex=False))
why_encoded['Power BI (표기 “Power Bi”, “파워비아이”, “powebi” 포함)'] = cond_pbi.astype(int)
```

```python
import pandas as pd

# 1. 결과를 담을 DataFrame 생성 (ID 포함)
incumbents_encoded = pd.DataFrame()
incumbents_encoded['ID'] = incumbents['ID']

# ==========================================
# A. 현직자 연차 (2개 컬럼)
# 대상 컬럼: incumbents_level
# ==========================================
# 규칙: "주니어" 포함 -> lvl_junior, "시니어" 포함 -> lvl_senior
incumbents_encoded['lvl_junior'] = incumbents['incumbents_level'].str.contains('주니어', na=False).astype(int)
incumbents_encoded['lvl_senior'] = incumbents['incumbents_level'].str.contains('시니어', na=False).astype(int)


# ==========================================
# B. 강연 내용 (3개 컬럼)
# 대상 컬럼: incumbents_lecture
# ==========================================
# 규칙: 키워드 포함 여부 확인 + "전부/모두" 같은 포괄적 응답 처리
target_col = incumbents['incumbents_lecture'].fillna('')

# 1) 개별 키워드 확인
has_trend = target_col.str.contains('산업 트렌드', regex=False)
has_career = target_col.str.contains('커리어 패스', regex=False)
has_job = target_col.str.contains('직무 강의', regex=False)

# 2) "전부", "모두", "3개" 등 모든 것을 포함하는 표현 확인 (Regex 사용)
has_all = target_col.str.contains('전부|모두|3개|3가지', regex=True)

# 3) 최종 컬럼 생성 (개별 키워드가 있거나 OR 전부 다라고 했으면 1)
incumbents_encoded['topic_industry_trend'] = (has_trend | has_all).astype(int)
incumbents_encoded['topic_career_path'] = (has_career | has_all).astype(int)
incumbents_encoded['topic_job_lecture'] = (has_job | has_all).astype(int)


# ==========================================
# C. 강연 방식 (2개 컬럼)
# 대상 컬럼: incumbents_lecture_type (내용상 이 컬럼이 온/오프라인 정보임)
# ==========================================
# 규칙: "온라인" 글자가 있으면 mode_online=1, "오프라인" 글자가 있으면 mode_offline=1
# "온, 오프라인 동시"의 경우 두 조건 모두 참이 되어 둘 다 1이 됨 (요청 사항 반영)
target_col = incumbents['incumbents_lecture_type'].fillna('')

incumbents_encoded['mode_online'] = target_col.str.contains('온라인', regex=False).astype(int)
incumbents_encoded['mode_offline'] = target_col.str.contains('오프라인', regex=False).astype(int)


# ==========================================
# D. 강연 규모 (3개 컬럼)
# 대상 컬럼: incumbents_lecture_scale
# ==========================================
# 규칙: 특정 인원 수 텍스트가 포함되어 있는지 확인
target_col = incumbents['incumbents_lecture_scale'].fillna('')

incumbents_encoded['scale_small'] = target_col.str.contains('10명 내외', regex=False).astype(int)
incumbents_encoded['scale_medium'] = target_col.str.contains('3~50명', regex=False).astype(int)
incumbents_encoded['scale_large'] = target_col.str.contains('100명 이상', regex=False).astype(int)
```

```python
import pandas as pd
import numpy as np

# 1. 필요한 열만 추출
df_major = major[['ID', 'major_field']].copy()

# 2. 분류 그룹 정의
high_tech_list = ['IT (컴퓨터 공학 포함)', '공학 (컴퓨터 공학 제외)']
data_friendly_list = ['경영학', '경제통상학', '자연과학', '자연고학']

def classify_majors(text):
    # 결측치 처리
    if pd.isna(text):
        return pd.Series([0, 0, 0])
    
    # 쉼표로 분리 및 공백 제거
    majors = [m.strip() for m in text.split(',')]
    
    high_tech = 0
    data_friendly = 0
    others = 0
    
    for m in majors:
        if m in high_tech_list:
            high_tech = 1
        elif m in data_friendly_list:
            data_friendly = 1
        else:
            # '없음'이나 빈 문자열이 아닌 경우만 Others로 카운트
            if m and m != 'nan':
                others = 1
                
    return pd.Series([high_tech, data_friendly, others])

# 3. 새로운 컬럼 적용 (ID 제외하고 3개 컬럼 생성)
df_major[['High Tech', 'Data Friendly', 'Others']] = df_major['major_field'].apply(classify_majors)

# 4. 최종 데이터프레임 구성
major_df = df_major[['ID', 'High Tech', 'Data Friendly', 'Others']]
```
