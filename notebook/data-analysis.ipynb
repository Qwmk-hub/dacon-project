```python
import numpy as np

# 각 열의 unique 값 분석
print("="*100)
print("각 열의 데이터 타입 및 Unique 값 분석")
print("="*100)

for col in data.columns:
    unique_count = data[col].nunique()
    null_count = data[col].isnull().sum()
    dtype = data[col].dtype
    
    print(f"\n📊 {col}")
    print(f"   데이터 타입: {dtype} | Unique 값 개수: {unique_count} | Null 개수: {null_count}")
    
    # unique 값이 20개 이하면 모두 출력, 초과면 샘플 출력
    unique_values = data[col].dropna().unique()
    
    if unique_count <= 20:
        print(f"   값: {sorted(unique_values) if dtype in ['int64', 'float64'] else sorted(unique_values)}")
    else:
        print(f"   샘플 값: {list(unique_values[:10])}")
        print(f"   ... (총 {unique_count}개의 unique 값)")

print("\n" + "="*100)
```

# 피처 엔지니어링

```python
# 데이터 재로드 (피처 엔지니어링용)
df_train = pd.read_csv("C:\\Users\\solba\\dacon-project\\data\\raw\\train.csv")
df_test = pd.read_csv("C:\\Users\\solba\\dacon-project\\data\\raw\\test.csv")

print("=" * 100)
print("피처 엔지니어링")
print("=" * 100)

# 불필요한 열 삭제 (generation, contest_award, idea_contest)
df_train = df_train.drop(columns=['generation', 'contest_award', 'idea_contest'])
df_test = df_test.drop(columns=['generation', 'contest_award', 'idea_contest'])

print("✅ 불필요한 열 삭제 (generation, contest_award, idea_contest)")

# 1. major_count: major1_1, major1_2 개수 (0, 1, 2)
df_train['major_count'] = df_train[['major1_1', 'major1_2']].notna().sum(axis=1)
df_test['major_count'] = df_test[['major1_1', 'major1_2']].notna().sum(axis=1)

# 2. is_IT: IT 여부
def check_it(row):
    m1 = str(row['major1_1']) if pd.notna(row['major1_1']) else ""
    m2 = str(row['major1_2']) if pd.notna(row['major1_2']) else ""
    return "IT(컴퓨터 공학 포함)" in m1 or "IT(컴퓨터 공학 포함)" in m2

df_train['is_IT'] = df_train.apply(check_it, axis=1)
df_test['is_IT'] = df_test.apply(check_it, axis=1)

print("✅ major_count, is_IT 생성")

# 3. class_count: 수강 수업 개수
df_train['class_count'] = df_train[['class1', 'class2', 'class3', 'class4']].notna().sum(axis=1)
df_test['class_count'] = df_test[['class1', 'class2', 'class3', 'class4']].notna().sum(axis=1)

print("✅ class_count 생성")

# 4. previous_class_count: 이전 기수 수강 (중복 제거)
def count_prev_classes(row):
    unique_classes = set()
    for col in ['previous_class_3', 'previous_class_4', 'previous_class_5', 
                'previous_class_6', 'previous_class_7', 'previous_class_8']:
        val = row[col]
        if pd.notna(val) and val != "해당없음":
            classes = [c.strip() for c in str(val).split(',')]
            unique_classes.update(classes)
    return len(unique_classes)

df_train['previous_class_count'] = df_train.apply(count_prev_classes, axis=1)
df_test['previous_class_count'] = df_test.apply(count_prev_classes, axis=1)

print("✅ previous_class_count 생성")

# 5. incumbents_lecture_scale_reason을 글자수로 변환
df_train['incumbents_lecture_scale_reason_len'] = df_train['incumbents_lecture_scale_reason'].apply(lambda x: len(str(x)) if pd.notna(x) else 0)
df_test['incumbents_lecture_scale_reason_len'] = df_test['incumbents_lecture_scale_reason'].apply(lambda x: len(str(x)) if pd.notna(x) else 0)

print("✅ incumbents_lecture_scale_reason_len 생성")
```

```python
# 6~10. 텍스트 나열형 변수를 개수로 변환

def count_items(val):
    """쉼표 단위로 항목 개수 세기. '없음'이거나 NaN이면 0"""
    if pd.isna(val):
        return 0
    val = str(val).strip()
    if val == "없음":
        return 0
    return len([x.strip() for x in val.split(',') if x.strip()])

# desired_job_count
df_train['desired_job_count'] = df_train['desired_job'].apply(count_items)
df_test['desired_job_count'] = df_test['desired_job'].apply(count_items)

# certificate_count ('없음' 처리)
df_train['certificate_count'] = df_train['certificate_acquisition'].apply(count_items)
df_test['certificate_count'] = df_test['certificate_acquisition'].apply(count_items)

# desired_certificate_count
df_train['desired_certificate_count'] = df_train['desired_certificate'].apply(count_items)
df_test['desired_certificate_count'] = df_test['desired_certificate'].apply(count_items)

# desired_job_except_data_count
df_train['desired_job_except_data_count'] = df_train['desired_job_except_data'].apply(count_items)
df_test['desired_job_except_data_count'] = df_test['desired_job_except_data'].apply(count_items)

# onedayclass_topic_count
df_train['onedayclass_topic_count'] = df_train['onedayclass_topic'].apply(count_items)
df_test['onedayclass_topic_count'] = df_test['onedayclass_topic'].apply(count_items)

print("✅ desired_job_count, certificate_count, desired_certificate_count, desired_job_except_data_count, onedayclass_topic_count 생성")
```

```python
# 원본 컬럼 삭제
drop_cols = ['major type', 'major1_1', 'major1_2', 'major_data',
             'class1', 'class2', 'class3', 'class4',
             'previous_class_3', 'previous_class_4', 'previous_class_5', 
             'previous_class_6', 'previous_class_7', 'previous_class_8',
             'desired_job', 'certificate_acquisition', 'desired_certificate', 
             'desired_job_except_data', 'onedayclass_topic',
             'incumbents_lecture_scale_reason', 'interested_company', 
             'expected_domain', 'completed_semester', 'major_field']

df_train = df_train.drop(columns=drop_cols)
df_test = df_test.drop(columns=drop_cols)

print("✅ 원본 컬럼 삭제 완료\n")

# 결과 확인
print("=" * 100)
print("최종 컬럼 목록:")
print("=" * 100)
print(df_train.columns.tolist())
print(f"\nTrain shape: {df_train.shape}")
print(f"Test shape: {df_test.shape}")
```

```python
# 데이터 타입 변경 (LabelEncoder 전에 수행)
print("=" * 100)
print("데이터 타입 변경")
print("=" * 100)

# int형으로 변환할 컬럼들 (카운트 및 수치형 피처)
int_columns = ['major_count', 'is_IT', 'class_count', 'previous_class_count', 
               'incumbents_lecture_scale_reason_len', 'desired_job_count', 
               'certificate_count', 'desired_certificate_count', 
               'desired_job_except_data_count', 'onedayclass_topic_count', 'time_input']

# int형으로 변환
for col in int_columns:
    df_train[col] = df_train[col].astype(int)
    df_test[col] = df_test[col].astype(int)

print(f"✅ int형으로 변환: {int_columns}")

# category형으로 변환할 컬럼들 (ID, completed, int_columns 제외한 나머지)
all_cols = df_train.columns.tolist()
category_columns = [col for col in all_cols if col not in ['ID', 'completed'] + int_columns]

# category형으로 변환
for col in category_columns:
    df_train[col] = df_train[col].astype('category')
    df_test[col] = df_test[col].astype('category')

print(f"✅ category형으로 변환: {category_columns}")
print("\n" + "=" * 100)
print("데이터 타입 변경 완료!")
print("=" * 100)
```

```python
# 나머지 범주형 변수들 LabelEncoder 적용 (completed 제외)
from sklearn.preprocessing import LabelEncoder

# 범주형 컬럼 리스트 (object 타입 + category 타입)
categorical_cols = df_train.select_dtypes(include=['object', 'category']).columns.tolist()

# ID, completed 제외 (completed는 타겟 변수, ID는 식별자)
if 'ID' in categorical_cols:
    categorical_cols.remove('ID')
if 'completed' in categorical_cols:
    categorical_cols.remove('completed')

print("=" * 100)
print("LabelEncoder 적용")
print("=" * 100)
print(f"대상 컬럼: {categorical_cols}\n")

# LabelEncoder 딕셔너리
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    
    # Train과 Test 데이터 합쳐서 fit (동일한 인코딩 규칙 적용)
    combined = pd.concat([df_train[col], df_test[col]])
    le.fit(combined)
    
    # Transform
    df_train[col] = le.transform(df_train[col])
    df_test[col] = le.transform(df_test[col])
    
    label_encoders[col] = le
    
    print(f"✅ {col}: {len(le.classes_)} classes")

print("\n" + "=" * 100)
print("LabelEncoder 완료!")
print("=" * 100)
```

```python
# is_IT를 0, 1로 변환 (이미 int형이지만 명시적으로)
df_train['is_IT'] = df_train['is_IT'].astype(int)
df_test['is_IT'] = df_test['is_IT'].astype(int)

print("✅ is_IT를 0, 1로 변환 완료")

# 열 이름 변경: ID, completed 제외하고 F1 ~ F32로 변경
cols = df_train.columns.tolist()

# ID, completed 제외한 피처 컬럼
feature_cols = [c for c in cols if c not in ['ID', 'completed']]

# 새로운 컬럼명 매핑 생성
column_mapping = {feature_cols[i]: f'F{i+1}' for i in range(len(feature_cols))}

# 컬럼명 변경
df_train = df_train.rename(columns=column_mapping)
df_test = df_test.rename(columns=column_mapping)

print(f"✅ 열 이름 변경 완료: F1 ~ F{len(feature_cols)}")

# completed를 y로 이름 변경하고 맨 마지막으로 이동
df_train = df_train.rename(columns={'completed': 'y'})

# 컬럼 순서 재정렬: ID, F1~F32, y
cols = df_train.columns.tolist()
cols.remove('y')
df_train = df_train[cols + ['y']]

print("✅ completed → y 로 변경 및 맨 마지막으로 이동")
print(f"최종 컬럼 순서: {df_train.columns.tolist()}")
```

```python
# 데이터 타입 최종 설정 (F1~F21: category, F22~F31: int)
print("=" * 100)
print("최종 데이터 타입 설정")
print("=" * 100)

# int형으로 유지할 컬럼 (F22~F31: 카운트 피처)
int_feature_list = [f'F{i}' for i in range(22, 32)]

# category형으로 변환할 컬럼 (F1~F21: 인코딩된 범주형 피처)
category_feature_list = [f'F{i}' for i in range(1, 22)]

# category형으로 변환
for col in category_feature_list:
    df_train[col] = df_train[col].astype('category')
    df_test[col] = df_test[col].astype('category')

print(f"✅ int형 컬럼 ({len(int_feature_list)}개): {int_feature_list}")
print(f"✅ category형 컬럼 ({len(category_feature_list)}개): {category_feature_list}")
print("\n" + "=" * 100)
print("최종 데이터 타입 설정 완료!")
print("=" * 100)
```
