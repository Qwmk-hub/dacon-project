```python
personal = train[['ID', 'school1', 'job', 'nationality', 'major type', 'major1_1', 'major1_2', 'major_field']]
```

```python
personal.head()
```

```python
high_tech_list = ['IT (컴퓨터 공학 포함)', '공학 (컴퓨터 공학 제외)']
data_friendly_list = ['경영학', '경제통상학', '자연과학', '자연고학']
def classify_majors(text):
    if pd.isna(text):
        return pd.Series([0, 0, 0])
    
    majors = [m.strip() for m in text.split(',')]
    
    high_tech = 0
    data_friendly = 0
    others = 0
    
    for m in majors:
        if m in high_tech_list:
            high_tech = 1
        elif m in data_friendly_list:
            data_friendly = 1
        else:
            if m and m != 'nan':
                others = 1
                
    return pd.Series([high_tech, data_friendly, others])

personal[['High Tech', 'Data Friendly', 'Others']] = personal['major_field'].apply(classify_majors)

personal_df = personal[['ID', 'school1', 'job', 'nationality', 'High Tech', 'Data Friendly', 'Others']]
```

```python
# 'certificate_acquisition' 열이 추가되었고, 여러 자격증이 콤마로 구분되어 있을 수 있음
# NaN, '해당없음', '없음' 제외, certificate_acquisition은 콤마로 분리하여 개수 합산
def count_valid(row):
    count = 0
    for col in row.index:
        if col == 'certificate_acquisition':
            val = row[col]
            if pd.notna(val) and val not in ['해당없음', '없음']:
                # 여러 자격증이 콤마로 구분되어 있을 수 있음
                count += len([x for x in str(val).split(',') if x.strip() and x.strip() not in ['해당없음', '없음']])
        elif col != 'ID':
            val = row[col]
            if pd.notna(val) and val not in ['해당없음', '없음']:
                count += 1
    return count

previous = train[['ID', 'class1', 'class2', 'class3', 'class4', 'previous_class_3', 'previous_class_4', 'previous_class_5',
                  'previous_class_6', 'previous_class_7', 'previous_class_8', 'certificate_acquisition']]
previous_count = previous.apply(count_valid, axis=1)
previous_count = pd.DataFrame({'ID': previous['ID'], 'count': previous_count})
```

```python
want_int = train[['ID', 'desired_career_path', 'desired_job', 'desired_job_except_data', 'desired_certificate', 'expected_domain']]
```

```python
# 원하는 것 개수 세기 함수 (expected_domain은 맨 앞 알파벳만 세기)
def count_want(row):
    count = 0
    for col in ['desired_job', 'desired_job_except_data', 'desired_certificate']:
        val = row[col]
        if pd.isna(val):
            continue
        items = [x.strip() for x in str(val).split(',') if x.strip() and x.strip() not in ['없음', '아직 없음']]
        count += len(items)
    # expected_domain은 맨 앞 알파벳만 세기
    val = row['expected_domain']
    if pd.notna(val):
        # 예: 'J. 정보통신업, O. 공공 행정, 국방 및 사회보장 행정' → ['J', 'O']
        items = [x.strip() for x in str(val).split(',') if x.strip()]
        domain_initials = [x.split('.')[0].strip() for x in items if '.' in x]
        count += len(domain_initials)
    return count

want_int['want_count'] = want_int.apply(count_want, axis=1)
want_int[['ID', 'want_count']].head()
```
